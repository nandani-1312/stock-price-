```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


try:
    df = pd.read_csv("TISC Historical Data.csv")
except FileNotFoundError:
    print("Error: File 'TISC Historical Data.csv' not found. Please make sure the file is in the correct directory.")
    exit()
except Exception as e:
    print(f"An error occurred while loading the data: {e}")
    exit()


try:
    df = df.rename(columns=lambda x: x.strip())
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
    df = df.sort_values('Date')

    df['Close'] = df['Price'].replace({',': ''}, regex=True)
    df['Close'] = pd.to_numeric(df['Close'], errors='coerce')


    if df['Close'].isnull().any():
        print("Warning: Some 'Price' values could not be converted to numeric. These have been set to NaN.")

    df = df.dropna(subset=['Close'])
except KeyError as e:
    print(f"Error: Key not found: {e}.  Please ensure the CSV file contains a column named 'Price'.")
    exit()
except Exception as e:
    print(f"An error occurred during data preprocessing: {e}")
    exit()

n = 5
for i in range(1, n + 1):
    df[f'Close_Lag_{i}'] = df['Close'].shift(i)


df.dropna(inplace=True)


X = df[[f'Close_Lag_{i}' for i in range(1, n + 1)]]
y = df['Close']


X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)


model = LinearRegression()
model.fit(X_train, y_train)


predictions = model.predict(X_test)


mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")


plt.figure(figsize=(12, 6))
plt.plot(df['Date'].iloc[-len(y_test):], y_test.values, label="Actual Price", linewidth=2)
plt.plot(df['Date'].iloc[-len(y_test):], predictions, label="Predicted Price", linewidth=2)
plt.legend()
plt.title("Tata Steel Stock Price Prediction vs Actual")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```
